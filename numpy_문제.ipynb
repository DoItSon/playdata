## 1. MAE 와 RMSE 함수 구현하기
- 다음과 같이 실제 값과 , 모델이 예측한 값이 각각 100개가 있다.
- 실제값과 예측값 사이에 평균 오차를 구하는 함수 두개를 구현해보자.

````python
np.random.seed(42) 
target = np.random.randint(50, 91,100) # 정답값
pred = np.random.randint(40, 100,100) # 예측값
target , pred
````

### MAE(Mean Absolute Error)
- 실제 값과 예측 값의 차이를 절대값으로 변환해 평균화
$$
MAE = \frac{1}{n}{\sum_{i=1}^{n}|y_i-\hat{y}_i|}
$$

```
아래 셀의 코드 작성하여 함수구현
```

````python
a = (np.abs(np.subtract(target,pred)))
sum(a)/100
````

````python
def mae(target,pred):
    a = (np.abs(np.subtract(target,pred)))
    return sum(a)/100
````

````python
mae(target,pred)
````

### RMSE(Root Mean Squared Error)
- 실제 값과 예측 값의 차이를 제곱한 뒤 평균화한 후에 루트

$$
RMSE = \sqrt{\frac{1}{n}{\sum_{i=1}^{n}(y_i-\hat{y}_i)^2}}
$$

```
아래 셀의 코드 작성하여 함수구현
```

````python
np.sqrt(np.divide(np.sum((np.subtract(target,pred))**2),100))
````

````python
def rmse(target,pred):
    return np.sqrt(np.divide(np.sum((np.subtract(target,pred))**2),100))
````

````python
rmse(target,pred)
````
