# Feature Engineering

- 새로 만든 feature 와 병합할 고객 번호로만 이루어진 데이터프레임을 만든다.

```python
features = pd.DataFrame(df["고객번호"].unique(),columns=["고객번호"])
features.head()
```

## 구매일자를 이용한 특성생성

- 내점일수
- 구매주기
- 주말방문비율
- 주구매요일
- 봄구매비율
- 여름구매비율
- 가을구매비율
- 겨울구매비율
- 일평균구매건수
- 거래개월수

```python
df["구매일자"].dt.weekday > 4
```

```python
(df["구매일자"].dt.weekday > 4).mean()
```

```python
agg_list = [
    # (컬럼명, 집계방식)
    ("내점일수","nunique"),
    ("구매주기",lambda x : (x.max() - x.min()).days / x.nunique() ), # 최근 구매일과 최초 구매일의 차이를 구하고 내점일수로 나누면 구매주기라고 볼수있음
    ("주말방문비율", lambda x : (x.dt.weekday > 4).mean() ), # 4보다 크면 토,일 이니까!!
    ("주구매요일" , lambda x: x.dt.weekday.mode()[0] ), # 최빈값이 여러개 일수도 있어서..
    ("봄구매비율" , lambda x : (x.dt.month.isin([3,4,5])).mean() ),
    ("여름구매비율" , lambda x : (x.dt.month.isin([6,7,8])).mean() ),
    ("가을구매비율" , lambda x : (x.dt.month.isin([9,10,11])).mean() ),
    ("겨울구매비율" , lambda x : (x.dt.month.isin([12,1,2])).mean() ),
    ("일평균구매건수" , lambda x : x.count() / x.nunique() ), # 구매건수를 구매일자로 나누면 일평균 구매건수라고 볼수있음.
    ("거래개월수" , lambda x : x.dt.month.nunique() )
 
]

tmp = df.groupby("고객번호")["구매일자"].agg(agg_list).reset_index()
features = features.merge(tmp,how="left",on = "고객번호") # features 변수에 merge 한 결과를 확인하고 변수에 넣으세요!! # 행이 늘어나거나 줄어들면 안된다.
features
```

## 지점을 이용한 특성생성
- 방문지점수
- 주구매지점

```python
agg_list = [
    # (컬럼명, 집계방식)
    ("방문지점수","nunique"),
    ("주구매지점" , lambda x : x.mode()[0])

]

tmp = df.groupby("고객번호")["지점"].agg(agg_list).reset_index()
features = features.merge(tmp,how="left",on = "고객번호") # features 변수에 merge 한 결과를 확인하고 변수에 넣으세요!!
features
```

## 상품번호를 이용한 특성생성
- 상품번호수

```python
agg_list = [
             ('상품번호수', 'nunique'),
             ]

tmp = df.groupby('고객번호')["상품번호"].agg(agg_list).reset_index()

features = features.merge(tmp, how='left') # features 변수에 merge 한 결과를 확인하고 변수에 넣으세요!!

features.head()
```

## 중분류를 이용한 특성생성
- 중분류수
- 주구매중분류

```python
agg_list = [
            ('중분류수', 'nunique'),
            ('주구매중분류', lambda x: x.mode()[0]), 
            ]

tmp = df.groupby('고객번호')["중분류"].agg(agg_list).reset_index()

features = features.merge(tmp, how='left')  # features 변수에 merge 한 결과를 확인하고 변수에 넣으세요!!

features.head()
```

## 대분류를 이용한 특성생성
- 대분류수
- 주구매대분류

```python
agg_list = [
            ('대분류수', 'nunique'),
            ('주구매대분류', lambda x: x.mode()[0]),
            ]

tmp = df.groupby('고객번호')["대분류"].agg(agg_list).reset_index()

features = features.merge(tmp, how='left')  # features 변수에 merge 한 결과를 확인하고 변수에 넣으세요!!

features.head()
```

## 구매가격을 이용한 특성생성
- 총구매액
- 구매건수
- 평균구매액
- 최대구매액
- 최소구매액
- 환불금액
- 환불건수
- 구매금액표준편차

```python
agg_list = [
        ('총구매액','sum'),
        ('구매건수', 'count'), 
        ('평균구매액', "mean"),
        ('최대구매액', 'max'),
        ('최소구매액',lambda x: x[x > 0].min() ) ,
        ('환불금액',lambda x: x[x < 0].sum() ) , # 마스킹 개념으로 True,False로 반환되어 수가 세어진다.
        ('환불건수', lambda x: ( x < 0 ).sum() ), # 마스킹 개념으로 True,False로 반환되어 True의 수가 세어진다.
        ('구매금액표준편차',lambda x: x[x>0].std() ),
    ]

tmp = df.groupby('고객번호')["구매가격"].agg(agg_list).reset_index()

features = features.merge(tmp, how='left')  # features 변수에 merge 한 결과를 확인하고 변수에 넣으세요!!

features.head()
```

```python
lst = ["농산물","가공식품","축산가공","수산품"]
agg_list = [
            ('식자재_구매수', lambda x: (x.isin(lst)).sum()), # count는 NaN데이터를 제외한 행의 수를 센다
            ]                                                   # sum은 True, False의 수를 센다.

tmp = df.groupby('고객번호')["대분류"].agg(agg_list).reset_index()
tmp
features = pd.merge(features,tmp,how="left",on="고객번호")
features
```

## 가설 2) 보석, 명품 등 사치용품(대분류)을 자주 사는 고객은 돈많은 여자일 것이다. (2007년 기준)

```python
df["대분류"].unique() # 사치용품 = 명품,보석,모피/피혁,액세서리,준보석/시계
```

```python
lst = ["명품","보석","모피/피혁","액세서리","준보석/시계"]
agg_list = [
            ('사치용품_구매수', lambda x: (x.isin(lst)).sum()), # count는 NaN 데이터를 제외한 행의 수를 센다
            ]
tmp = df.groupby('고객번호')["대분류"].agg(agg_list).reset_index()
features = pd.merge(features,tmp,how="left",on="고객번호")
features
```

## 가설 3) 거래 개월 수에 작고 내점일수가 클 때 자주 방문함으로 여성일 것이다. 따라서 거래개월수 * 내점일수 수치가 클수록 여성일 확률이 크다

```python
agg_list = [
    # (컬럼명, 집계방식)
    ("거래개월수X내점일수" , lambda x :x.nunique() * x.dt.month.nunique())
]

tmp = df.groupby("고객번호")["구매일자"].agg(agg_list).reset_index()
features = features.merge(tmp,how="left",on = "고객번호")
features
```

## 가설 4) 평일환불고객은 여자일 것이다.
## 2007년 시대 상황보면 전업주부가 많았을 것으로 예상된다.
## 그래서 평일에 환불하려면 평일에 시간이 상대적으로 많은 여성이 많을 것이다.

```python
def do_apply(x):
    mask = x["구매일자"].dt.weekday < 5
    return (x.loc[mask, "구매가격"]<0).sum()

tmp = df.groupby("고객번호").apply(do_apply)
tmp = tmp.reset_index()
tmp = tmp.rename(columns = {0:"평일환불수"})

features = features.merge(tmp,how="left",on = "고객번호")
features
```
