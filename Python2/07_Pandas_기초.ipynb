# 행렬 다루기
- 데이터프레임에는 행에 대한 이름을 index라고 하고, 열에 대한 이름을 column 이라 한다.
- 데이터프레임도 numpy 기반으로 돌아가기 때문에, 행번호와 열번호가 있다.

- `iloc`
    - 행번호, 열번호를 이용한 행렬 슬라이싱
- loc[0],iloc[0]의 차이점은 loc는 전체 프레임에서 인덱스 이름이 0인 행만 출력하고,
iloc[0] 전체 프레임에서 0번쨰 행에 있는 값들만 추출해라!

```python
df2.head()
```

```python
df2.iloc[:5,1:5] 
```

```python
num_rows = [1,3]
num_cols = [0,4,5]
df2.iloc[num_rows,num_cols]
```

```python
df2.iloc[:3,::-1]
```

- `loc`
    - index 명과 column 명을 이용한 슬라이싱
    - 마스킹이용한 행과 열 선택이 가능하다.

```python
df.loc[:3] # end 번호까지 포함
```

```python
df.loc[:,["embarked","age"]]
```

```python
idx = [0,2,3]
cols = ["name","age","parch"]
df.loc[idx,cols]
```

- 다중 조건을 주어 마스킹
    - 데이터프레임에서는 논리 연산자를 지원하지 않는다.
    - `or` 대신 `|`(Vertical bar)
        - 둘중에 하나가 1 이면 1을 반환
    - `and` 대신 `&`(Ampersand)
        - 둘다 1이면 1을 반환
    - `not` 대신 `~`(tilde)
        - 1 면 0 을반환
    `

```python
True == 1 , False == 0
```

```python
mask1 = df2["target"] == 1
mask2 = df2["age"] < 20
df2.loc[mask1 & mask2].head()
```

- map 메소드 사용해보기
    - 딕셔너리를 전달해서 컬럼의 값들을 변경할수 있다.
    - `key` 는 변경전 값, `value` 는 변경후 값

```python
replace_dict = {"male":1,"female":0}
df["gender"].map(replace_dict)
```

```python
df["gender"].map(lambda x: 1 if x == "male" else 0) # 비율이 적은 쪽에 1을 넣음
```
