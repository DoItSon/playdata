## 마스킹 (masking)
- bool 배열을 마스크로 사용하여 데이터의 특정 부분을 선택할 수 있다.

```python
arr
```

```python
mask_list = [True,False,True,False] # 행 선택
arr[mask_list]
```

```python
mask_list = [True,False,True] # 열 선택
arr[:,mask_list]
```

```python
mask = arr > 2
mask
```

```python
arr[mask] # 모든 스칼라에 마스킹을 할 경우 벡터로  반환된다.
```

- Norm
    - 벡터의 크기를 측정하는 함수
$$
L_1 = {\sum_{i=1}^{n}|x_i|}
$$
$$
L_2 = \sqrt{{\sum_{i=1}^{n}x_i^2}}
$$

```python
np.linalg.norm([-1,2,3],1) # L1 norm
```

```python
np.linalg.norm([-1,2,3],2) # L2 norm (2는 디폴트라 입력안해도 된다!)
```

- 내적(dot product)
    - 두 벡터의 각 요소끼리 곱의 합
    - 결과값은 스칼라가 나온다.

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
np.dot(a,b)
```

```python
a @ b # 이렇게도 쓰인다!
```

- 행렬곱
    - 2차원 공간에서 내적한다.
    - 앞에 행렬의 열개수와 뒤에 행렬의 행개수가 동일해야한다.
    - 연산결과는 shape는 앞에 행렬의 행개수와 뒤에 행렬의 열개수가 나온다.

```python
a = np.array([3,2,2]) / np.linalg.norm([-1,2,3],2)# 사과, 딸기, 참외의 빈도
b = np.array([0,1,2]) / np.linalg.norm([-1,2,3],2) # 문서의 유사도를 구분
np.dot(a,b)
```

- `shuffle`
    - 반환값이 없고, 배열 자체를 섞는다.
    - 원본 데이터가 섞여서 위험성이 높다.
    - arr = arr.copy (깊은 복사를 사용해야한다.)

- `choice`
